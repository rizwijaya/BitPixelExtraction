```python
#Library yang akan digunakan
import numpy as np
from PIL import Image
from google.colab import files
```

```python
#Fungsi untuk mendapatkan RGB per Pixel
def get_valueRGB(image):
  image_data = np.asarray(image)
  print("RGB per pixel : \n")
  for i in range(len(image_data)):
    for j in range(len(image_data[0])):
      print(image_data[i][j])
```

```python
#Fungsi untuk mendapatkan bitdepth
def get_bitdepth(image, m): 
  bpp = {'RGB'  == m or 'YCbCr' == m : '24',
         'RGBA'  == m or 'CMYK' == m: '32'}.get(
             True, 'Tidak ada yang sesuai')
  print("Bit Depth = ", bpp) #Cetak bitdeptnya
```

```python
uploaded = files.upload() #Upload image
for fn in uploaded.keys():
  path_img = fn
image = Image.open(path_img)
im = image.mode
print(im) #Mencetak mode gambar
display(image) #Menampilkan gambar
```

```python
get_bitdepth(image, im) #Panggil fungsi bitdepth
```

```python
get_valueRGB(image) #Panggil fungsi RGB per Pixel
```
