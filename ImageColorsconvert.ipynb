```python
#Library yang akan digunakan
import numpy as np
from PIL import Image
from google.colab import files
import pyhdust as hdt
import pyhdust.images as phim
```

```python
#Fungsi konversi rgb to cmyk
def rgbtocmyk(img):
  img = phim.rgb2cmyk(np.asarray(img))
  img = Image.fromarray(np.asarray(img), mode='CMYK')
  img.save('cmyk.jpeg') #Menyimpan gambar hasil konversi cmyk
  #Menampilkan gambar CMYK
  print(img.mode)
  #imgcmyk = Image.open('cmyk.jpg')
  #display(imgcmyk)
```

```python
#Fungsi rgb to ycrcb
def rgbtoycrcb(img2):
  img2 = Image.fromarray(np.asarray(img2), mode='YCbCr')
  img2.save('YcrCb.jpeg')
  #Menampilkan gambar hasil konversi rgb to YcrCb
  print(img2.mode)
  imgYcrCb = Image.open('YcrCb.jpeg')
  display(imgYcrCb)
```

```python
#Form Upload Images
uploaded = files.upload()
for fn in uploaded.keys():
  path = fn
imgori = Image.open(path)
imgori = imgori.convert(mode='RGB')
```

```python
#Menampilkan gambar sebelum convert
print(imgori.mode)
display(imgori)
```

```python
rgbtoycrcb(imgori) #Memanggil fungsi konversi RGB to YCRCB
```

```python
rgbtocmyk(imgori) #Memanggil fungsi konversi RGB to CMYK
```
